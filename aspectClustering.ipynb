```
#leitura do arquivo csv e montagem da string
def readCsv2str():
  sample = open("laptop_filtered_aspect_sample.csv", "r") 
  data = sample.read()
  data = data.replace(" ","_")
  data = data.replace("\n"," ")
  #desconsiderando o título da coluna (aspect name)
  data = data[12:]
  return data

```

```
#função realiza a leitura do arquivo csv e retorna uma lista das sentenças
def readCsv2List(name):
  with open(name, 'r') as f:
    reader = csv.reader(f)
    your_list = list(reader)
  flat_list = [item for sublist in your_list for item in sublist]
  size = len(flat_list)
  #cria uma lista de n = size listas
  lista = [[] for _ in range(size)]
  #o índice é utilizado desse jeito para ignorar o nome da coluna (aspect name)
  for i  in range(1,size):
    lista[i-1].append(flat_list[i])
  return lista
```

```
lista = readCsv2List("laptop_filtered_aspect_sample.csv")
#treinanamento dos modelos, variando o algoritmo
model1 = Word2Vec(lista, min_count=1,size= 100,workers=3, window = 2, sg = 1)
model2 = Word2Vec(lista, min_count=1,size= 100,workers=3, window = 2, sg = 0)
```

```
def KMEANS(k,X):
  kclusterer = KMeansClusterer(k, distance=nltk.cluster.util.cosine_distance, repeats=25)
  assigned_clusters = kclusterer.cluster(X, assign_clusters=True)
  return assigned_clusters
```

```
#função responsável por remover a pontuação e 
def preProcessingCorpus(reviews):

  stop_words = set(stopwords.words('english')) 
  #criando uma lista com n = len(reviews) listas
  filtered_sentence = [[] for i in range(len(reviews))]
  for i in range(len(reviews)):
    words = re.findall(r'\w+', str(reviews[i]),flags = re.UNICODE) 
    for word in words:
      if word not in stop_words:
          filtered_sentence[i].append(word)
  return filtered_sentence
```

```
#download de corpus de reviews de produtos e de stopwords
nltk.download('product_reviews_1')
nltk.download('product_reviews_2')
nltk.download('stopwords')

reviews = product_reviews_1.sents()
reviews2 = product_reviews_2.sents()

corpus1 = preProcessingCorpus(reviews)
corpus2 = preProcessingCorpus(reviews2)
corpus = corpus1 + corpus2
```

```
#treinamento do terceiro modelo com o corpus préprocessado
model3 = gensim.models.Word2Vec(corpus, min_count=1,size= 100,workers=5, window = 10, sg = 1)
```
